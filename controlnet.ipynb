```python
def save_dataset_to_folder(folder_path, ds):
    # Создание дополнительных папок
    os.makedirs(folder_path + 'images', exist_ok=True)
    os.makedirs(folder_path + 'conditioning_images', exist_ok=True)
    os.makedirs(folder_path + 'text', exist_ok=True)
    
    # Сохранение изображений
    for i, example in enumerate(ds):
        image = example['image']
        conditioning_image = example['conditioning_image'] 
        image.save(os.path.join(folder_path + 'images/', f'image_{i}.jpg'))
        conditioning_image.save(os.path.join(folder_path + 'conditioning_images/', f'image_{i}.jpg'))
        
    # Сохранение текстовых описаний в формате json
    text_json = [example['text'] for example in ds]
    with open(folder_path + 'text.json', "w") as f:
        json.dump(text_json, f, indent=4)

save_dataset_to_folder('/kaggle/working/ControlNet/training/', ds)   
```

```python
class MyDataset(Dataset):
    def __init__(self):
        self.data = []
        with open('/kaggle/working/ControlNet/training/text.json', 'rt') as f:
            self.texts = json.load(f)

        for i in range(len(self.texts)):
            self.data.append(
                {'image': f'images/image_{i}.jpg',
                 'conditioning_image': f'conditioning_images/image_{i}.jpg',
                 'text': self.texts[i]}
            ) 

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]

        source_filename = item['conditioning_image']
        target_filename = item['image']
        prompt = item['text']

        source = cv2.imread('/kaggle/working/ControlNet/training/' + source_filename)
        target = cv2.imread('/kaggle/working/ControlNet/training/' + target_filename)

        if source is None:
            raise FileNotFoundError(f"Source image not found: {source_filename}")
        if target is None:
            raise FileNotFoundError(f"Target image not found: {target_filename}")

        source = cv2.cvtColor(source, cv2.COLOR_BGR2RGB)
        target = cv2.cvtColor(target, cv2.COLOR_BGR2RGB)

        source = source.astype(np.float32) / 255.0

        target = (target.astype(np.float32) / 127.5) - 1.0

        return dict(jpg=target, txt=prompt, hint=source)
```

```python
class ImageLoggerWithWandb(ImageLogger):
    def __init__(self, *args, save_dir="/kaggle/working/logs", yadisk_token=None, yadisk_path="/logs", **kwargs):
        super().__init__(*args, **kwargs)
        self.save_dir = save_dir
        self.yadisk_token = yadisk_token
        self.yadisk_path = yadisk_path
        self.yadisk_client = None
        
        if yadisk_token is not None:
            self.yadisk_client = YaDisk(token=yadisk_token)
            if not self.yadisk_client.check_token():
                raise ValueError("Invalid Yandex.Disk token")
        
    def log_local(self, save_dir, split, images, global_step, current_epoch, batch_idx):
        effective_save_dir = save_dir if save_dir is not None else self.save_dir
        
        super().log_local(effective_save_dir, split, images, global_step, current_epoch, batch_idx)
        
        wandb_logs = {}
        if hasattr(self, "logger") and self.logger:
            for k in images:
                grid = torchvision.utils.make_grid(images[k], nrow=4)
                
                if self.rescale:
                    grid = (grid + 1.0) / 2.0
                
                grid = grid.cpu().numpy().transpose(1, 2, 0)
                grid = (grid * 255).astype(np.uint8)
                
                wandb_logs[f"images/{k}"] = wandb.Image(grid)
                
                # Сохранение изображения на диск
                image_pil = Image.fromarray(grid)
                image_filename = os.path.join(
                    effective_save_dir, 
                    f"{split}_epoch{current_epoch}_step{global_step}_batch{batch_idx}_{k}.png"
                )
                image_pil.save(image_filename)
                
                # Загрузка на Яндекс.Диск
                if self.yadisk_client is not None:
                    try:
                        remote_path = os.path.join(
                            self.yadisk_path,
                            f"{split}_epoch{current_epoch}_step{global_step}_batch{batch_idx}_{k}.png"
                        )
                        remote_dir = os.path.dirname(remote_path)
                        
                        # Создание директории на Яндекс.Диске, если её нет
                        if not self.yadisk_client.exists(remote_dir):
                            self.yadisk_client.mkdir(remote_dir, recursive=True)
                        
                        # Загрузка файла
                        self.yadisk_client.upload(image_filename, remote_path)
                    except Exception as e:
                        print(f"Error uploading to Yandex.Disk: {str(e)}")
            
            self.logger.experiment.log(wandb_logs)
```

```python
# Инициализация логгера для визуализации в W&B
logger_wb = WandbLogger(project="controlnet")

# Основные параметры обучения
resume_path = '/kaggle/working/ControlNet/SD/control_sd15_ini.ckpt'
batch_size = 2
logger_freq = 300
learning_rate = 1e-5
sd_locked = True
only_mid_control = False

# Инициализация модели
model = create_model('/kaggle/working/ControlNet/models/cldm_v15.yaml')
model.load_state_dict(load_state_dict(resume_path, location='cpu'))
model.learning_rate = learning_rate
model.sd_locked = sd_locked
model.only_mid_control = only_mid_control

# Настройка даталоадера
dataset = MyDataset()
dataloader = DataLoader(dataset, num_workers=0, batch_size=batch_size, shuffle=True, pin_memory=True)

# Логгер для сохранения примеров изображений
logger = ImageLoggerWithWandb(
    batch_frequency=500,
    max_images=4,
    rescale=True,
    save_dir="/kaggle/working/image_log",
    yadisk_token=YANDEX_TOKEN,
    yadisk_path="Папочка/Images"
)
```

```python
BUFFER_DIR = '/kaggle/working/checkpoints_temp'
REMOTE_DIR = "Папочка/Checkpoints"

os.makedirs(BUFFER_DIR, exist_ok=True)
```

```python
class YandexDiskUploader:
    def __init__(self, buffer_dir: str, yadisk_token: str, remote_dir: str, max_buffer_size_gb: float = 1.0):
        self.buffer_dir = buffer_dir
        self.remote_dir = remote_dir if remote_dir.startswith("/") else "/" + remote_dir
        self.y = YaDisk(token=yadisk_token)
        self.stop_event = threading.Event()
        self.max_buffer_size = max_buffer_size_gb * 1024**3

        if not self.y.check_token():
            raise ValueError("Invalid Yandex.Disk token")
        if not self.y.exists(self.remote_dir):
            self.y.mkdir(self.remote_dir, recursive=True)

        self.thread = threading.Thread(target=self._monitor_buffer, daemon=True)
        self.thread.start()

    @retry(stop_max_attempt_number=3, wait_fixed=2000)
    def _safe_upload(self, local_path: str):
        remote_filename = os.path.basename(local_path)
        remote_path = f"{self.remote_dir}/{remote_filename}"
        print(f"→ Загружаем {local_path} → {remote_path}")
        if not self.y.exists(remote_path):
            self.y.upload(local_path, remote_path)
            print("✔ Uploaded successfully")
        else:
            print("⚠ Уже существует, пропускаем")

    def _cleanup_buffer(self):
        files = sorted(
            (os.path.join(self.buffer_dir, f) for f in os.listdir(self.buffer_dir)),
            key=os.path.getmtime
        )
        total = sum(os.path.getsize(f) for f in files)
        while total > self.max_buffer_size and files:
            os.remove(files.pop(0))
            total = sum(os.path.getsize(f) for f in files)

    def _monitor_buffer(self):
        print("Uploader thread started, buffer_dir =", self.buffer_dir)
        while not self.stop_event.is_set():
            try:
                self._cleanup_buffer()
                for fname in os.listdir(self.buffer_dir):
                    if fname.endswith(".ckpt"):
                        local = os.path.join(self.buffer_dir, fname)
                        self._safe_upload(local)
                        os.remove(local)
                        print(f"Удалён локально: {fname}")
            except Exception as e:
                print(f"Uploader error: {e}")
                time.sleep(5)
            time.sleep(10)

    def stop(self):
        self.stop_event.set()
        self.thread.join()


# -------- SETUP ----------

# 1) Запускаем поток загрузки на Яндекс.Диск
uploader = YandexDiskUploader(
    buffer_dir=BUFFER_DIR,
    yadisk_token=YANDEX_TOKEN,
    remote_dir=REMOTE_DIR,
    max_buffer_size_gb=7
)

# 2) Создаем callback для сохранения чекпоинтов модели
checkpoint_callback = ModelCheckpoint(
    dirpath=BUFFER_DIR,
    filename="model_epoch={epoch:02d}_step={global_step}",  
    save_top_k=-1,
    save_weights_only=True,
    every_n_train_steps=500,
    verbose=True
)

# 3) Callback для progress bar 
progress_callback = TQDMProgressBar(refresh_rate=20)

# 4) Создаем trainer
trainer = pl.Trainer(
    gpus=1,
    max_epochs=1,
    callbacks=[checkpoint_callback, progress_callback, logger],
    logger = logger_wb,
    num_sanity_val_steps=0,
    enable_progress_bar=True
)

# 5) Запуск обучения
try:
    trainer.fit(model, dataloader)
finally:
    uploader.stop()
    for f in os.listdir(BUFFER_DIR):
        os.remove(os.path.join(BUFFER_DIR, f))
```
